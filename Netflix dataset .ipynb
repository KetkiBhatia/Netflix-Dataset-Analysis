```python
Q2. In which year the highest number of TV shows and Movies were released? Show with the bar graph.
```

```python
data.dtypes
```

```python
data['Date_N'] = pd.to_datetime(data['release_year'])
```

```python
data.head()
```

```python
data['Date_N'].dt.year.value_counts()
```

```python
import pandas as pd

# Assuming data is your DataFrame

# Count the occurrences of each release year
year_counts = data['release_year'].value_counts()

# Display the counts sorted by year
year_counts_sorted = year_counts.sort_index()
print(year_counts_sorted)
```

```python
# Ensure the release_year column is of integer type
data['release_year'] = data['release_year'].astype(str)

# Create the Date_N column using the release_year column
data['Date_N'] = pd.to_datetime(data['release_year'], format='%Y', errors='coerce')

# Extract the year from the Date_N column
data['Year'] = data['Date_N'].dt.year

# Count the occurrences of each year
year_counts = data['Year'].value_counts().sort_index()

print(year_counts)

max_release_year = year_counts.idxmax()
max_releases = year_counts.max()

print(f"The year with the maximum number of releases is {max_release_year} with {max_releases} releases.")
```

```python
data['Date_N'].dt.year.value_counts().sort_index().plot(kind='bar', figsize=(12, 6), color='skyblue')
import matplotlib.pyplot as plt

threshold = 100
filtered_year_counts = year_counts[year_counts >= threshold]

# Sort the filtered counts from high to low
filtered_year_counts = filtered_year_counts.sort_values(ascending=False)

# Plot the bar graph
filtered_year_counts.plot(kind='bar', figsize=(12, 6), color='skyblue')
# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.title('Number of Releases per Year')
plt.xticks(rotation=45)
plt.show()
```
